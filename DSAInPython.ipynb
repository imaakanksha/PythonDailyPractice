Container with most water Problem
Given n non-negative integers where each integer represent the hight of a vertical line on a chart.
Find two lines, which together with x-axis forms a container, that holds the biggest amount of water. return the area of that water.

index  heights
0       5
1       9
2       2
3       1
4       4
5       3


```python
def maxArea(self, height: list[int]) -> int:
    maxarea=0
    l = 0
    r =len(height)-1
    while(l<r):
        maxarea=max(maxarea, min(height[l], height[r]) * (r-l))
        if height[l] < height[r]:
            l+=1
        else:
            r-=1
    return maxarea
```

Two Sum
Given an array of integers nums and an integer target, return indices of the two numbers such that they add up to target.
You may assume that each input would have exactly one solution, add you may no use the same element twice. You can return the answer in any order

nums = [2,7,11,15] target=9
nums = [3,3] target =6
nums = [3,1,7,9,2,4] target = 6

```python
#class Solution:
def twoSum(self, nums: list[int], target:int) -> list[int]:
        hashmap={}
        for i in range(len(nums)):
            complement = target - nums[i]
            
            if complement in hashmap:
                return[hashmap[complement], i]
            hashmap[nums[i]] = i
```

What are decorators in Python? 
Decorators in Python are a design pattern that allows you to add new functionality to an existing object without modifying its structure. They are commonly used to extend the behavior of functions or methods. You can read more about how to use Python decorators in a separate guide. 

```python
import functools

def my_decorator(func):
    @functools.wraps(func)          # preserves __name__, __doc__, etc.
    def wrapper(*args, **kwargs):
        print("Something is happening before the function is called.")
        result = func(*args, **kwargs)
        print("Something is happening after the function is called.")
        return result
    return wrapper

@my_decorator
def say_hello():
    print("Hello!")

say_hello()
# Output:
# Something is happening before the function is called.
# Hello!
# Something is happening after the function is called.
```

Valid Mountain Array
Given an array of integers, return true if the following conditions are fulfilled.
Length of the array is bigger than or equal to 3
There exists some index i such that:
a[0] < a[1] < ....< a[i]
a[i] > a[i+1] ...> a[a.size-1]
In a nutshell: Find if there is an increasing subarray followed by a decreasing sub array
 [3,5,5] => output ; False. It is not a vaild mountain array
 [0,3,2,1] => output ; True. It is a vaild mountain array

```python
def validMountainArray(self, A: list[int]) -> bool:
        i = 1
        while(i<len(A) and A[i] > A[i-1]):
            i+=1
        if(i==1 or i==len(A)):
            return False
        while(i<len(A) and A[i] < A[i-1]):
            i+=1
        return i == len(A)
```

Contains Duplicate
Given an integer array nums, return true if any value appears at least twice in the array, and return false if every element is distinct
[1,2,3,1] => output: True
[5,6,7,8,9] => output: False

```python
def containsDuplicate(self, nums: list[int]) -> bool:
    hashset = set()
    for num in nums:
        if num in hashset:
            return True
        hashset.add(num)

    return False

#Time Complexity : O(1)
#Space Complexity : O(n)
```
